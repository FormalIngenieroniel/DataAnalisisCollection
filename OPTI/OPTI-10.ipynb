```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
salesDist = pd.read_csv('stores-dist.csv')
salesDist.head()
```

```python
salesDist.rename(columns={"annual net sales":"sales","number of stores in district":"stores"}, inplace = True)
salesDist
```

```python
salesDist.drop(["district"],axis = 1, inplace = True)
salesDist.head()
```

```python
y = salesDist['sales']
x = salesDist['stores']
```

```python
%matplotlib inline 

plt.figure(figsize=(20,10))

plt.plot(x,y,'o',markersize = 15)

plt.ylabel('Annual Net Sales', fontsize = 30)
plt.xlabel('Number of Stores in District', fontsize = 30)

plt.xticks(fontsize = 30)
plt.yticks(fontsize = 30)

plt.show()
```

```python
m, b= np.polyfit(x,y,1)
print('The slope of line is {:.2f}.'.format(m))
print('The y-intercept is {:.2f}.'.format(b))
print('The best fit simple linear regression line is {:.2f}x +{:.2f}.'.format(m,b))
```

```python
y_mean = y.mean()
x_mean = x.mean()
print('Centroide x = {:.2f} and y = {:.2f}.'.format(x_mean, y_mean))
```

```python
%matplotlib inline 

plt.figure(figsize=(20,10))
plt.plot(x,y,'o',markersize = 14, label ="Annual Net Sales")

plt.plot(x_mean,y_mean,'*',markersize = 30, color = "r")
plt.plot(x, m*x +b ,'-',label = 'Simple Linear Regr', linewidth = 4)

plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)

plt.annotate('Centroid',xy=(x_mean-0.1,y_mean-5), xytext=(x_mean-3,y_mean-20), arrowprops=dict(facecolor='black',shrink=0.05), fontsize = 30)

plt.legend(loc='upper right',fontsize = 20)

print("AQUI",m,b)
```

```python
def predict(query):
  if query >= 1:
    predict = m* query + b
    return predict
  else:
    print("lO ULTIMO")
```
